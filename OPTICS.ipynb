```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:,3:5].values
```

```python
from sklearn.cluster import OPTICS
```

```python
OP = OPTICS(min_samples=2).fit(x)
```

```python
y_OP=OP.fit_predict(x)
```

```python
superviced = dataset
```

```python
superviced
```

```python
superviced['Cluter_group'] = y_OP
```

```python
superviced
```

```python
import seaborn as sns
fecet = sns.lmplot(data=superviced, x=superviced.columns[3] , y=superviced.columns[4],hue=superviced.columns[1],fit_reg=False,legend=True,  legend_out=True)
```
